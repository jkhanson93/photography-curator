```python
import os
from PIL import Image, ExifTags
import piexif
```

```python
photo = Image.open("C:/Users/jkhan/Code/Python/PhotographyCurator/photography-curator/photos/DSCF5001.JPG")
exif_data = piexif.load(photo.info.get("exif", b""))
exif = exif_data.get("Exif", {})
print(exif)

for tag_id, value in exif.items():
    tag_name = ExifTags.TAGS.get(tag_id)
    print(f"{tag_name}: {value}")
```

```python
date_taken = exif.get(piexif.ExifIFD.DateTimeOriginal)
iso = exif.get(piexif.ExifIFD.ISOSpeedRatings)
focal_length = exif.get(piexif.ExifIFD.FocalLength)
```

```python
print(f"Date Taken: {date_taken.decode()}")
print(f"ISO: {iso}")
print(f"Focal Length: {focal_length[0]/focal_length[1]}mm")
```
